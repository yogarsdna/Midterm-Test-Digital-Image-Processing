## Gaussian Blur

```python
#Import the libraries that will be used
from PIL import Image
from PIL import ImageFilter

#Open the image
img = Image.open("pedri2.jpg")

#Blur the image using Gaussian Blur
blur_img =img.filter(ImageFilter.GaussianBlur(5))

#Show the image that has been blurred
blur_img.show()
```

```python
#Import the libraries that will be used
from skimage import io, filters

#Open the image
img = io.imread("luffy.jpg")

#Blur the image using Gaussian Blur
out = filters.gaussian(img, sigma = 5)

#Show the image that has been blurred
io.imshow(out)
```

## Median Filter

```python
#Import the libraries that will be used
from PIL import Image
from PIL import ImageFilter

#Open the image
img = Image.open("luffy.jpg") 

#Blur the image using median filter with the size of 7
blur_img = img.filter(ImageFilter.MedianFilter(7))

#Show the image that has been blurred
blur_img.show()
```

## Dilation

```python
#Import the libraries that will be used
from skimage import morphology
from skimage import io, color
from skimage.morphology import erosion, dilation, opening, closing, white_tophat, disk
from skimage.util import img_as_ubyte

#Open the image
img = io.imread("luffy.jpg")

#Convert the image to a gray image
gray = color.rgb2gray(img)

#Show the image where the color has been converted to gray
io.imshow(gray)
```

```python
#Return grayscale morphological dilation of the image
img_filter = img_as_ubyte(gray)
dilated = dilation(img_filter, disk(3))

#Show the image that has been dilated
io.imshow(dilated)
```

## Rotation

```python
#Import the libraries that will be used
import cv2

#Read and get the shape of the image
img = cv2.imread('nishinoya.jpg',0)
rows,cols = img.shape

#Get the rotation matrix using cv2.getRotationMatrix2D
M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)

#Rotate the image using cv2.warpAffine 
dst = cv2.warpAffine(img, M, (cols, rows))

#Show the image that has been rotated
cv2.imshow('img', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Scaling

```python
#Import the libraries that will be used
import cv2
import numpy as np

#Read the image
img = cv2.imread('nishinoya.jpg')

#Specify the scaling factor
res = cv2.resize(img, None, fx = 1.5, fy = 1.5, interpolation = cv2.INTER_CUBIC)

#Show the image that has been scaled
cv2.imshow('img',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Translation

```python
#Import the libraries that will be used
import numpy as np
import cv2 as cv

#Read and get the shape of the image
img = cv.imread('nishinoya.jpg',0)
rows,cols = img.shape

#Create the transformation matrix M
M = np.float32([[1, 0, 100],[0, 1, 50]])

#Applies an affine transformation to the image
dst = cv.warpAffine(img, M, (cols, rows))

#Show the image that has been shifted
cv.imshow('img', dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

## Affine Transformation

```python
#Import the libraries that will be used
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Read and get the shape of the image
img = cv.imread('nishinoya.jpg')
rows, cols, ch = img.shape

#Make three points from input and output image to find the transformation matrix
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])

#Create a 2x3 matrix from pts1 and pts2
M = cv.getAffineTransform(pts1, pts2)

#Applies an affine transformation to the image
dst = cv.warpAffine(img, M, (cols, rows))

#Plot the initial image and the image that has been given the Affine Transformation
plt.subplot(121), plt.imshow(img) ,plt.title('Input')
plt.subplot(122), plt.imshow(dst) ,plt.title('Output')
plt.show()
```

## Perspective Transformation

```python
#Import the libraries that will be used
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Read and get the shape of the image
img = cv.imread('nishinoya.jpg')
rows,cols,ch = img.shape

#Make three points from input and output image to find the transformation matrix
pts1 = np.float32([[56, 65], [368, 52], [28, 387],[389, 390]])
pts2 = np.float32([[0, 0], [300, 0], [0, 300], [300, 300]])

#Create a 2x3 matrix from pts1 and pts2
M = cv.getPerspectiveTransform(pts1, pts2)

#Applies a perspective transformation to the image
dst = cv.warpPerspective(img, M, (300, 300))

#Plot the initial image and the image that has been given the Perspective Transformation
plt.subplot(121), plt.imshow(img), plt.title('Input')
plt.subplot(122), plt.imshow(dst), plt.title('Output')
plt.show()
```
